```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
data = pd.read_csv('job-market(2).csv')
```

```python
%matplotlib inline

dataEdit = data['Classification'].value_counts().sort_values(ascending=False)
print(dataEdit.head())
sns.barplot( y=dataEdit.index.to_list(), x=dataEdit)
```

```python
import matplotlib
#create dataframe for salary ranges
dataSalary = data['LowestSalary'].value_counts()
print(dataSalary)
```

```python
#salary ranges plot
fig = plt.figure(figsize=(8,8))
dataLabels = ['0-30','30-40','40-50','50-60']

plt.pie(dataSalary, labels=dataLabels,autopct='%1.1f%%')
```

```python
locationDf = data[data['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf['Count'] = locationDf["Id"]
locationDf.head()
```

```python
%matplotlib inline
# Pie char for job market share of cities
from matplotlib import cm
plt.pie(locationDf['Count'][:7], labels=locationDf['Location'][:7],autopct='%1.1f%%')
```
